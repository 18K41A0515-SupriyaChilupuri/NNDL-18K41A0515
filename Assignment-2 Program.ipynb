```python
from sympy import Symbol, Derivative, symbols
x= Symbol('x')
y= Symbol('y')
function = x**2 + y**2 + 10
partialderivative= Derivative(function, y)
dfy = partialderivative.doit()
partialderivative= Derivative(function, x)
dfx = partialderivative.doit()
x1 = -1.0 
y1 = 2.0
itr = 100  
learningrate = 0.1 
for i in range(0,itr):
    x = symbols('x')
    y = symbols('y')
    dfxv = dfx.subs(x, x1) 
    dfxv = round(dfxv,2)
    dfyv = dfy.subs(y, y1) 
    dfyv = round(dfyv,2)
    dx = (-1.0)*learningrate*dfxv
    dy = (-1.0)*learningrate*dfyv
    x1 = x1 + dx
    x1 = round(x1, 2) 
    y1 = y1 + dy
    y1 = round(y1, 2)
print(x1, y1)
print(f'min value obtained is x = {x1} ,y ={y1} for given function')
```
